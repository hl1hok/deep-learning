```python
with tf.device('/gpu:0'):
    def next_batch(batch_size, image, label):
        
        a = np.random.randint(np.shape(image)[0] -batch_size)
        batch_x = image[a:a + batch_size, :]
        batch_y = label [a:a + batch_size, :]
        return batch_x, batch_y
```
